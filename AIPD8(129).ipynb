```python
def detect_adversarial(sample):

    noise = np.random.normal(0, 0.05, sample.shape)

    perturbed = sample + noise

    pred1 = model.predict(sample.reshape(1, -1))
    pred2 = model.predict(perturbed.reshape(1, -1))

    if pred1 != pred2:
        return "⚠️ Possible Adversarial Attack"
    else:
        return "✅ Safe Input"
```
